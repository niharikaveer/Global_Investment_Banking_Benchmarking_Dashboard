```python
import yfinance as yf
import pandas as pd

banks = {
    "HLI": "Houlihan Lokey",
    "EVR": "Evercore",
    "JEF": "Jefferies Financial Group",
    "SF": "Stifel Financial",
    "GS": "Goldman Sachs",
    "MS": "Morgan Stanley",
    "JPM": "JPMorgan Chase",
    "LAZ": "Lazard",
    "RJF": "Raymond James Financial",
    "MC": "Moelis & Company"
}

for ticker, bank_name in banks.items():

    print(f"Downloading {ticker}...")

    df = yf.download(
        ticker,
        start="2021-01-01",
        end="2026-01-01",
        auto_adjust=False,
        progress=False
    )

    df = df.reset_index()

    # Keep only required columns
    df = df[[
        "Date",
        "Open",
        "High",
        "Low",
        "Close",
        "Volume"
    ]]

    # Rename columns
    df.columns = [
        "Date",
        "Open_Price",
        "High_Price",
        "Low_Price",
        "Close_Price",
        "Volume"
    ]

    # Add bank information
    df["Ticker"] = ticker
    df["Bank_Name"] = bank_name

    # Arrange columns
    df = df[
        [
            "Date",
            "Ticker",
            "Bank_Name",
            "Open_Price",
            "High_Price",
            "Low_Price",
            "Close_Price",
            "Volume"
        ]
    ]

    # Save individual CSV
    filename = f"{ticker}_stock_prices.csv"

    df.to_csv(filename, index=False)

    print(f"Saved {filename}")

print("\nAll CSV files downloaded successfully!")
```
